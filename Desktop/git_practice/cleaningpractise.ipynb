```python
import pandas as pd
dummy_df =pd.DataFrame({'name': ["Isaac","Catherine","Tafford","Sandra"],
                        'score':[20,34,56,98,]
                        })
dummy_df['score'] = dummy_df['score'].map(lambda x: x-5)
```

```python
print(dummy_df)
```

```python
dummy_df['length'] = dummy_df['name'].apply(len)
print(dummy_df)
```

```python
dummy_df = dummy_df.applymap(lambda x:x * 2)
print(dummy_df)
```

```python
dummy_df =pd.DataFrame({'name': ["Isaac Macharia","Catherine Nanjala","Tafford Pessah","Sandra Bosibori"]
                        
                        })
dummy_df['last_name'] = dummy_df['name'].apply(lambda x: x.split()[1])
print(dummy_df)
```

```python
dummy_df =pd.DataFrame({'email': ["IsaacMacharia@gmail.com","CatherineNanjala@gmail.com","TaffordPessah@gmail.com","SandraBosibori@gmail.com"]
                        
                        })
dummy_df['first_name'] = dummy_df['email'].apply(lambda x: x.split('@')[0])
print(dummy_df)
```

```python
dummy_df =pd.DataFrame({'email': ["Isaac.Macharia@gmail.com","Catherine.Nanjala@gmail.com","Tafford.Pessah@gmail.com","Sandra.Bosibori@gmail.com"]
                        
                        })
dummy_df['first_name'] = dummy_df['email'].apply(lambda x: x.split('@')[0])
dummy_df['last_name'] = dummy_df['email'].apply(lambda x: x.split('@')[1])

print(dummy_df)
```

```python
dummy_df['last_name1'] = dummy_df['first_name'].apply(lambda x: x.split('.')[0])

dummy_df['last_name2'] = dummy_df['first_name'].apply(lambda x: x.split('.')[1])

print(dummy_df)
```
